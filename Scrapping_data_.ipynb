```python
import pandas as pd
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Create an empty DataFrame to store data
product_data = pd.DataFrame(columns=["ID", "Name", "Price", "Brand"])
# Unique ID counter
unique_id = 1

for index_num in range(5):

    if (index_num == 1):
        buttons = driver.find_element(By.CLASS_NAME, 'box--eTFFd')
        button = buttons.find_element(By.XPATH,'/html/body/div[3]/div/div[3]/div/div/div[1]/div[3]/div/div/ul/li[3]')
        button.click()
    elif (index_num == 2):
        buttons = driver.find_element(By.CLASS_NAME, 'box--eTFFd')
        button = buttons.find_element(By.XPATH,
                                      '/html/body/div[3]/div/div[3]/div/div/div[1]/div[3]/div/div/ul/li[4]')
        button.click()
    elif (index_num == 3):
        buttons = driver.find_element(By.CLASS_NAME, 'box--eTFFd')
        button = buttons.find_element(By.XPATH,'/html/body/div[3]/div/div[3]/div/div/div[1]/div[3]/div/div/ul/li[5]')
        button.click()
    elif(index_num ==4):
        buttons = driver.find_element(By.CLASS_NAME, 'box--eTFFd')
        button = buttons.find_element(By.XPATH,'/html/body/div[3]/div/div[3]/div/div/div[1]/div[3]/div[2]/div/ul/li[6]')
        button.click()
    
    index_num += 1

    for index in range(40):
        ad_grid = driver.find_element(By.XPATH, '//*[@id="root"]/div/div[3]/div/div/div[1]/div[2]')
        ad_cards = ad_grid.find_elements(By.CLASS_NAME, 'gridItem--Yd0sa')

        try:
            # Click on the current ad card
            current_card = ad_cards[index]
            current_card.click()

            Information_grid = driver.find_element(By.XPATH, '/html/body/div[4]/div/div[3]')

            # Extract only name, price, and brand
            name_of_product = Information_grid.find_element(By.XPATH, '//*[@id="module_product_title_1"]/div/div')
            product_name = name_of_product.text
            prices = Information_grid.find_elements(By.CLASS_NAME, 'pdp-product-price')
            price = prices[0].text

            # Extract brand information
            brand_element = Information_grid.find_element(By.XPATH, '/html/body/div[4]/div/div[3]/div[2]/div/div[1]/div[5]/div/a[1]')
            brand = brand_element.text

            # Append data to DataFrame with unique ID and brand
            product_data = product_data.append({"ID": unique_id, "Name": product_name, "Price": price, "Brand": brand},
                                              ignore_index=True)

            # Increment unique ID
            unique_id += 1

        except Exception as e:
            print(f"Error processing product {index + 1}: {str(e)}")
        driver.back()

# Save data to a CSV file
product_data.to_csv("product_data.csv", index=False)

# Print completion message
print("Data scraped and saved successfully!")
```
