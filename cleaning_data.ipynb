```python
import pandas as pd
import re

# Read the CSV file into a DataFrame
csv_file_path = 'product_data_with_cleaned_columns.csv'
df = pd.read_csv(csv_file_path)

# Function to clean and extract the device name from the 'Name' column
def clean_and_extract_device_name(name):
    # Convert to lowercase and remove specific words and characters
    cleaned_name = name.lower().replace('xiaomi', '').replace('||', '').replace('-', '').replace(',', '').replace('ram','').replace('|','')

    # Remove occurrences of "gb" followed by a number
    cleaned_name = re.sub(r'\d+gb', '', cleaned_name)

    # Extract the first three words
    words = cleaned_name.split()[:3]
    if words[2] not in ['12', 'x5', '12s','12S','pro']:
        words = words[0:2]
    return ' '.join(words)

# Apply the function to create the 'Device Name' column
df['Device Name'] = df['Name'].apply(clean_and_extract_device_name)


df.to_csv('product_data_with_cleaned_d.csv', index=False)
```

## Storing Data into DataBase

```python
import sqlite3
import pandas as pd

# Read the CSV file into a DataFrame
csv_file_path = 'z_mobile_data_cleaned_2.csv'
df = pd.read_csv(csv_file_path)

# Create a new SQLite database
conn = sqlite3.connect('new_product_data.db')
cursor = conn.cursor()

# Create a Products table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS Products (
        ID INTEGER PRIMARY KEY,
        Name TEXT,
        Price TEXT,
        Brand TEXT,
        Rating TEXT,
        Total_Ratings TEXT,
        Cleaned_Price INTEGER,
        Clean_Rating REAL,
        Brand_Name TEXT
    )
''')

# Iterate through the DataFrame and insert data into the Products table
for _, row in df.iterrows():
    # Insert data into the Products table
    cursor.execute('''
        INSERT INTO Products (ID, Name, Price, Brand, Rating, Total_Ratings, Cleaned_Price, Clean_Rating, Brand_Name)
        VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)
    ''', (
        row['ID'], row['Name'], row['Price'], row['Brand'], row['Rating'], row['Total Ratings'],
        row['Cleaned_Price'], row['Clean_Rating'], row['Brand Name']
    ))

# Commit changes and close the connection
conn.commit()
conn.close()
```

## Cleaning out other data sheet

```python
import pandas as pd

# Read the CSV file into a DataFrame
csv_file_path = 'product_data_with_cleaned_d.csv'
df = pd.read_csv(csv_file_path)

# Function to filter devices based on user query
def get_best_devices_within_price_range(min_price, max_price):
    # Convert price columns to numeric
    df['Cleaned_Price'] = pd.to_numeric(df['Cleaned_Price'], errors='coerce')

    # Filter devices within the specified price range
    filtered_df = df[(df['Cleaned_Price'] >= min_price) & (df['Cleaned_Price'] <= max_price)]

    # Sort devices by rating and price
    sorted_df = filtered_df.sort_values(by=['Clean_Rating', 'Cleaned_Price'], ascending=[False, True])

    # Extract relevant information for the top 2 devices
    result = sorted_df[['ID', 'Device Name', 'Cleaned_Price', 'Clean_Rating', 'Total Ratings']].head(2)

    return result

# Prompt the user to enter a price range
min_price = float(input("Enter the minimum price: "))
max_price = float(input("Enter the maximum price: "))

# Get the best devices based on the user's price range
result = get_best_devices_within_price_range(min_price, max_price)

# Display the result
print(f"Based on user ratings and price, the best phones in the price range of {min_price} to {max_price} PKR are:")
print(result)
```

### Adding Ram

```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('.csv')

# Function to extract RAM information or assign a random value if not available
def extract_ram(row):
    ram_str = str(row['Name'])  
    if 'GB RAM' in ram_str:
        return int(ram_str.split('GB')[0].strip())
    else:
        return 4  # Assigning a default value of 4GB if RAM information is not available

# Apply the function to create a new 'RAM' column
df['RAM'] = df.apply(extract_ram, axis=1)

# Write the updated DataFrame to a new CSV file
df.to_csv('your_updated_csv_file.csv', index=False)
```
